# Data cleaning before merging the dataframes


```python
#while going through the trial dataset .. we found that there is a co-relation between the columns 
#('LOSS' and 'LOSS_SCORE') and ('HISTORY','HISTORY_SCORE').. So we decided to drop one of the columns from the dataset.
#We first converted the LOSS_SCORE as per the correlation into a new column STATUS_LOSS and then compared the values to 
#confirm that both columns are same. and similarly we did it with HISTORY_SCORE .  in which we found that 6 records needed 
# imputation . So as per the HISTORY_SCORE we imputed the data in column HISTORY.
```

```python
trial_df.HISTORY.unique()
```

```python
trial_df['HISTORY'].value_counts()
```

```python
# checking observations corrosponding to  HISTORY : 3,9,5,4

history_value = [3,9,5,4]

trial_df.loc[trial_df['HISTORY'].isin(history_value)]
```

```python
### as the history score for these value is same i.e 6 , we changed the value of the HIStory column to 2, which is the corresponding code
# for HISTORY_SCORE 6.. also changing the same in audit_df

trial_df.loc[trial_df.HISTORY == 3, 'HISTORY'] = 2
trial_df.loc[trial_df.HISTORY == 5, 'HISTORY'] = 2
trial_df.loc[trial_df.HISTORY == 9, 'HISTORY'] = 2
trial_df.loc[trial_df.HISTORY == 4, 'HISTORY'] = 2

audit_df.loc[audit_df.HISTORY == 3, 'HISTORY'] = 2
audit_df.loc[audit_df.HISTORY == 5, 'HISTORY'] = 2
audit_df.loc[audit_df.HISTORY == 9, 'HISTORY'] = 2
audit_df.loc[audit_df.HISTORY == 4, 'HISTORY'] = 2
```

```python
#checking if the data contains these values
history_value = [3,9,5,4]

trial_df.loc[trial_df['HISTORY'].isin(history_value)]
```

```python
#Converting the HISTORY_SCORE into a new column STATUS_HISTORY according to the relation found between HISTORY_SCORE
#and HISTORY. and checking if all values are same.
trial_df.loc[trial_df.HISTORY_SCORE == 2, 'STATUS_HISTORY'] = 0  
trial_df.loc[trial_df.HISTORY_SCORE == 4, 'STATUS_HISTORY'] = 1  
trial_df.loc[trial_df.HISTORY_SCORE == 6, 'STATUS_HISTORY'] = 2 
RESULT_HISTORY=trial_df.HISTORY == trial_df.STATUS_HISTORY
RESULT_HISTORY.value_counts()
```

```python

from sklearn.linear_model import Ridge

x_range = [0.01, 0.1, 1, 10,100]
train_score_list = []
test_score_list = []

for alpha in x_range: 
    ridge = Ridge(alpha)
    ridge.fit(X_train,y_train)
    train_score_list.append(ridge.score(X_train,y_train))
    test_score_list.append(ridge.score(X_test, y_test))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x_range, train_score_list, c = 'g', label = 'Train Score')
plt.plot(x_range, test_score_list, c = 'b', label = 'Test Score')
plt.xscale('log')
plt.legend(loc = 3)
plt.xlabel(r'$\alpha$')
```

``Appropriate value for alpha as per graph 10``

```python
import math

print("log alpha array: {}".
      format(np.around(np.log(np.array(x_range)),3)))
print("alpha array: {}".format(np.array(x_range)))
print("")

for i in [-2,-1,0,1]:
    print("Training performance when alpha is ", math.pow(10, i) ,": {:.5f}".format(train_score_list[i+2]))
    print("Test performance when alpha is ", math.pow(10, i) ,": {:.5f}".format(test_score_list[i+2]))
    print("")

print("Ridge Regression Coefficients: \n\n",ridge.coef_)
print("")
print("Ridge Regression intercept value: ",ridge.intercept_)
print("")
```

``Appropriate value for alpha in Ridge Regression is",10 as it has the lowest difference between train and test scores``

```python

from sklearn.model_selection import cross_val_score
# search for an optimal value of K for KNN

# list of integers 1 to 30
# integers we want to try
k_range = range(1, 31)

# list of scores from k_range
k_scores = []

# 1. we will loop through reasonable values of k
for k in k_range:
    # 2. run KNeighborsRegressor with k neighbours
    knn = KNeighborsRegressor(n_neighbors=k)
    # 3. obtain cross_val_score for KNeighborsRegresor with k neighbours
    scores = cross_val_score(knn,X_train_org, y_train, cv=10)
    # 4. append mean of scores for k neighbors to k_scores list
    k_scores.append(scores.mean())
print(k_scores)
```

```python
print("Average cross-validation score: {:.2f}".format(scores.mean()))
```

```python
# plot the value of K for KNN (x-axis) versus the cross-validated accuracy (y-axis)
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
```

From the graph we can see thatthe model gives best cross-validated acccuracy for K=9
